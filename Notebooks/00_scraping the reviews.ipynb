```python
import pandas as pd
# ^^^ pyforest auto-imports - don't write above this line
import warnings
warnings.filterwarnings("ignore")
from selenium import webdriver
from selenium.webdriver.common.by import By
import math

chromedriver_path = "../chromedriver.exe"
chrome_options = webdriver.ChromeOptions()
driver = webdriver.Chrome(executable_path=chromedriver_path)
```

```python
Base_url = "https://www.amazon.com/Fitpolo-Waterproof-Monitoring-Pedometer-Activity/product-reviews/B07Y4PT2NH/ref=cm_cr_dp_d_show_all_btm?ie=UTF8&reviewerType=all_reviews"


# get the number of pages 
def get_pages_no(url):
    chromedriver_path = "../chromedriver.exe"
    chrome_options = webdriver.ChromeOptions()
    driver = webdriver.Chrome(executable_path=chromedriver_path)
    driver.get(url)
    x = driver.find_element(By.XPATH,'//*[@id="filter-info-section"]/div')
    number = x.text.split(" ")[-3].replace(",","")
    return math.ceil(int(number)/10)


# Name of customer
Names = []

# Review content
Review = []


# get number of pages 
no_of_pages = get_pages_no(Base_url)


for i in range(0,no_of_pages):
    full_review_info = []
    url = Base_url+"&pageNumber="+str(i)
    driver.get(url)
    # Slice the section that contains all reviews on web page
    reviews_section = driver.find_element(By.XPATH, '//*[@id="cm_cr-review_list"]')

    # take reviews_section section and divided it into individual reviews sections
    sections_for_review = reviews_section.find_elements(By.CLASS_NAME, 'review')

    # Lists used to store

    
    for i in sections_for_review:
        full_review_info.append(i)

    

    # the review content 
    for i in full_review_info:
        x = i.find_element(By.CLASS_NAME ,'a-section').text
        x = x.split("\n")[4:-2]
        Review.append(" ".join(x))
        
    
    # Get the name of customer that make review
    for i in full_review_info:
        x = i.find_element(By.CLASS_NAME, 'a-profile-name')
        Names.append(x.text)
```

```python
df = pd.DataFrame(list(zip(Names, Review)),
               columns =['Customer Name', 'Customer review'])
```

```python
df.to_csv("Reviews.csv")
```
